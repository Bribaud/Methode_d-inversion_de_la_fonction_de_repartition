```r
# Fonction pour gÃ©nÃ©rer des Ã©chantillons alÃ©atoires selon la loi exponentielle Exp(Î»)
generate_exponential <- function(n, lambda) {
  # GÃ©nÃ©rer des Ã©chantillons alÃ©atoires selon une distribution uniforme sur [0, 1]
  u <- runif(n)
  # Appliquer la mÃ©thode de l'inverse de la fonction de rÃ©partition
  x <- -log(1 - u) / lambda
  return(x)
}

# ParamÃ¨tres de la loi exponentielle
lambda <- 0.5  # ParamÃ¨tre lambda

# GÃ©nÃ©rer 1000 Ã©chantillons alÃ©atoires selon la loi exponentielle Exp(Î»)
samples <- generate_exponential(1000, lambda)

# Afficher les Ã©chantillons alÃ©atoires gÃ©nÃ©rÃ©s
hist(samples, breaks = 30, col = "lightblue", main = "Histogramme de la loi exponentielle", xlab = "Valeurs", ylab = "FrÃ©quence")
```

## Loi de Cauchy

```r
# Fonction pour gÃ©nÃ©rer des Ã©chantillons alÃ©atoires selon la loi de Cauchy
generate_cauchy <- function(n, lambda) {
  # GÃ©nÃ©rer des Ã©chantillons alÃ©atoires selon une distribution uniforme sur [0, 1]
  u <- runif(n, 0, 1)
  # Appliquer la mÃ©thode de l'inverse de la fonction de rÃ©partition
  x <- lambda * tan(pi * (u - 0.5))
  return(x)
}

# ParamÃ¨tres de la loi de Cauchy
lambda <- 1  # ParamÃ¨tre lambda

# GÃ©nÃ©rer 1000 Ã©chantillons alÃ©atoires selon la loi de Cauchy
samples_cauchy <- generate_cauchy(1000, lambda)

# Afficher un histogramme des Ã©chantillons gÃ©nÃ©rÃ©s
hist(samples_cauchy, breaks = 30, col = "lightgreen", main = "Histogramme de la loi de Cauchy", xlab = "Valeurs", ylab = "FrÃ©quence")
```

## MÃ©thode de Box-Muller

1Â°)

```r
# Nombre d'Ã©chantillons Ã  gÃ©nÃ©rer
n <- 1000

# GÃ©nÃ©rer les Ã©chantillons de X_k suivant une loi uniforme sur [-1, 1]
X <- runif(n, min = -1, max = 1)

# Calculer la somme cumulÃ©e de X_k
cumulative_sum <- cumsum(X)

# Calculer les valeurs de la variable alÃ©atoire de la somme normalisÃ©e
normalized_sum <- cumulative_sum / sqrt(1:n)

# CrÃ©er un histogramme pour reprÃ©senter les rÃ©alisations de la somme normalisÃ©e
hist(normalized_sum, breaks = 30, col = "lightblue", main = "Histogramme de la somme normalisÃ©e", xlab = "Valeurs de la somme normalisÃ©e")
```

```r
# Nombre d'Ã©chantillons Ã  gÃ©nÃ©rer
n <- 1000

# GÃ©nÃ©rer les Ã©chantillons de X_k suivant une loi de Cauchy avec lambda = 1
X <- rcauchy(n, location = 0, scale = 1)

# Calculer la somme cumulÃ©e de X_k
cumulative_sum <- cumsum(X)

# Calculer les valeurs de la variable alÃ©atoire de la somme normalisÃ©e
normalized_sum <- cumulative_sum / sqrt(1:n)

# CrÃ©er un histogramme pour reprÃ©senter les rÃ©alisations de la somme normalisÃ©e
hist(normalized_sum, breaks = 30, col = "lightblue", main = "Histogramme de la somme normalisÃ©e", xlab = "Valeurs de la somme normalisÃ©e")
```
